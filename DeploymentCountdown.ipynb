```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from datetime import date
start_date = date(2022, 2, 1)
end_date = date(2022, 7, 10)
delta = end_date - start_date
#print("Length of Deployment ") 
#print(delta)

days_completed = date.today() - start_date
print("Days Completed ")
print(days_completed)

days_remaining = end_date - date.today()
print("Days Remaining ")
print(days_remaining)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Days Completed', 'Days Remaining'
sizes = [days_completed / delta, days_remaining / delta]
explode = (0.1,  0)  # only "explode" the 1st Slice (days completed)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=["green", "red"], autopct='%1i%%',
        shadow=True, startangle=135)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
#plt.title = ("Main Body % Deployment Complete", bbox={'facecolor':'0.8', 'pad':5})
ax1.set_title("Main Body % Deployment Complete")
plt.show()
```
